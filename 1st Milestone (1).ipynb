```python
import pandas as pd

file_path = "data.csv"
df = pd.read_csv("Downloads/Compressed/Expanded_data_with_more_features.csv")
df
```

```python
print(df[['MathScore', 'ReadingScore', 'WritingScore']].head())
```

```python
columns_to_average = ['MathScore', 'ReadingScore', 'WritingScore']

df['AverageScore'] = round(df[columns_to_average].sum(axis=1) / len(columns_to_average))
print(df[['AverageScore']].head())
```

```python
df['TotalPerformance'] = df['AverageScore'].apply(
    lambda x: 'high' if x > 85 else 'medium' if x >= 70 else 'low'
)
print(df[['TotalPerformance']].head())
```

```python
df.drop(columns=['AverageScore'], inplace=True)
print(df.head())
```

```python
df_cleaned = df_modified.copy()

df_cleaned = df_cleaned.drop(columns=["EthnicGroup"])

print(df_cleaned.head())
```

```python
from sklearn.impute import SimpleImputer


df_imputed = df_cleaned.copy()

num_cols = df_imputed.select_dtypes(include=['float64', 'int64']).columns
num_imputer = SimpleImputer(strategy='median')
df_imputed[num_cols] = num_imputer.fit_transform(df_imputed[num_cols])


cat_cols = df_imputed.select_dtypes(include=['object']).columns
cat_imputer = SimpleImputer(strategy='most_frequent')
df_imputed[cat_cols] = cat_imputer.fit_transform(df_imputed[cat_cols])

missing_after = df_imputed.isnull().sum().to_dict()
missing_after
```

```python
df_imputed = df_imputed.drop(columns=["Unnamed: 0"])
```

```python
df_imputed = df_imputed.drop(columns=["Gender"])
```

```python
import numpy as np

def convert_hours(value):
    if pd.isna(value):
        return np.nan
    value = str(value).strip()
    
    if value.startswith("<"):
        return float(value[1:]) / 2   # نصف القيمة (مثلاً <5 = 2.5)
    elif "-" in value:
        parts = value.split("-")
        return (float(parts[0]) + float(parts[1])) / 2   # المتوسط
    elif value.startswith(">"):
        return float(value[1:]) + 1   # أكبر من الرقم +1 (مثلاً >10 = 11)
    else:
        try:
            return float(value)
        except:
            return np.nan

df_imputed["WklyStudyHours_num"] = df_imputed["WklyStudyHours"].apply(convert_hours)
```
